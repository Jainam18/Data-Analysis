# Univariate Feature Selection

Univariate Feature Selection works by selecting the best features based on univariate statistical tests. It can be seen as a preprocessing step to an estimator

```python
import pandas as pd
import numpy as np #for linear algebra
import seaborn as sns #data visualization library
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE #RFE stands for recursive feature elimination
```

```python
br = pd.read_csv('../Dataset/breast.csv')
br.head()
```

```python
y = br.diagnosis #diagnosis response variable
```

# Drop the irrelevant columns

```python
list = ['Unnamed: 32','id','diagnosis']  #list of columns we dont need
```

```python
x = br.drop(list,axis=1) #dropping columns 
x.head()
```

# Univariate Feature Selection

Use SelectKBest that removes all but the k highest scoring features

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#find best scored 5 features
selected_feature = SelectKBest(chi2, k=5).fit(x, y)
```

```python
print(selected_feature)
```

```python
print('Score List:', selected_feature.scores_)
print('Feature list:', x.columns)
```

# Recursive Feature Elimination (RFE)

RFE with Random Forest Algorithm. 
It uses one of the classification methods (random forest), assign weights to each features (Xs with smallest weights are removed)

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfe = RFE(estimator=rfc, n_features_to_select=5, step=1)
rfe = rfe.fit(x,y)
```

```python
print('Chosen best 5 feature by rfe', x.columns[rfe.support_])
```

# Principal Component Analysis (PCA)

It is a dimentionality reduction technique. 

```python
from sklearn.decomposition import PCA
from sklearn .preprocessing import scale
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
num = iris[["sepal_length","sepal_width","petal_length","petal_width"]]
```

```python
#convert numeric values to numpy arrays - also called as standardization
X = num.values
X
```

```python
#Scaling the values
X = scale(X)
X
```

```python
pca = PCA(n_components=4) # create a covariance matrix in which we have 4 features
```

```python
#Calculate Eigenvalues
pca.fit(X)
```

```python
#The amount of variance that each PC explains
var = pca.explained_variance_ratio_
var
```

```python
#Cumulative Variance explained with [n] features
var1 = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
plt.plot(var1)
plt.ylabel('% Variance Explained')
plt.xlabel('# of Features')
plt.title('PCA Analysis')
```
